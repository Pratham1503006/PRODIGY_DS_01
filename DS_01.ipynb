## Extarction of top 10 countries with respect to total population

```python
file_path = "C:/Users/pkg15/OneDrive/Desktop/Internship/Task 01/API_SP.POP.TOTL_DS2_en_csv_v2_9949.csv"
df = pd.read_csv(file_path, skiprows=4)

print(df.columns)

total_population_data = df[df['Indicator Code'] == 'SP.POP.TOTL']
total_population_sorted = total_population_data.sort_values(by="2022", ascending=False)

total_top_ten_countries = total_population_sorted.head(10)

print("Top ten countries of total population\n")
print(total_top_ten_countries[['Country Code', '2022']])
```

# Bar Plot

## Top ten countries of total population in year 2022

```python
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2022)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

## Extraction of bottom 10 countries with respect to total popultaion

```python
total_population_sorted1 = total_population_data.sort_values(by="2022", ascending=True)


total_bottom_ten_countries = total_population_sorted1.head(10)
print("Bottom ten countries of total population\n")
print(total_bottom_ten_countries[['Country Code']])
```

## Bottom ten countries of total population in year 2016

```python
plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2016", y="Country Code", data=total_bottom_ten_countries, palette="coolwarm")
plt.title("Bottom Ten Countries of Total Population (2016)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
sorted_data = total_population_sorted
top_10_countries = sorted_data.head(10)
plt.figure(figsize=(12,6))
sns.barplot(x="2022",y="Country Code",data=top_10_countries,color="red",label="2022")
sns.barplot(x="1987",y="Country Code",data=top_10_countries,color="green",label="1987")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sea
import pandas as pd

file_path = "C:/Users/pkg15/OneDrive/Desktop/Internship/Task 01/API_SP.POP.TOTL_DS2_en_csv_v2_9949.csv"
df = pd.read_csv(file_path, skiprows=4)
country_name = input("Enter the name of the country: ")
country_data = df[df['Country Name'] == country_name]  
year = country_data.columns[4:]
population = country_data.iloc[0, 4:].values  
plt.figure(figsize=(20, 10))  # Corrected figure size
plt.plot(year, population, marker='o', markersize=5) 
plt.title(f'Population Growth of {country_name}') 
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
